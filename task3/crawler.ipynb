```python
import requests
from bs4 import BeautifulSoup

resp = requests.get('https://www.slickcharts.com/market-movers', 
                    headers={'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36'})

market_info = {}
if resp.status_code == 200:
    soup = BeautifulSoup(resp.text, "html.parser")

    target_divs = soup.find_all('div', class_ = 'shadow p-3 mb-4 bg-white rounded')

    for div in target_divs:
        h5_tag = div.find('h5')
        if h5_tag:
            # find the market index name
            market_index = h5_tag.find('a')
            # stocks in the market
            stock_block = div.find_all('td', class_ = "text-nowrap")

            if market_index:
                market_info[market_index.text] = {}               
                stock_names = []
                
                # find stock name
                for stock in stock_block:
                    flag = 0
                    stock_name = stock.find('a')
                    
                    if stock_name:
                        stock_names.append(stock_name.text)                                   

                    # find stock percentage        
                    if '(' in stock.text:
                        percentage = stock.text.replace("(", "").replace(")","")  
                        flag = 1

                    # update the dictionary
                    if flag == 1:
                        name = stock_names.pop(0)
                        market_info[market_index.text][name] = percentage
                        continue     
                       
for key, item in market_info.items():
    print(key)
    for subkey, subitem in item.items():
        print(subkey, subitem)
    print("\n")
```
